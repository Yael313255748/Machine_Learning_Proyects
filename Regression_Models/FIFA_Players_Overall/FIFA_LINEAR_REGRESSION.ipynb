```python
import numpy as np
import pandas as pd
from tensorflow.keras import optimizers,constraints,losses,callbacks
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import pickle
from random import seed
from tensorflow.random import set_seed
from Machine_Learning_Tools.Neural_Networks_Models import single_layer_perceptron
warnings.filterwarnings('ignore')
%matplotlib inline
```

# **Preprocesamiento de los datos y elección de variables**

```python
fifa_20 = pd.read_csv('datasets/players_20.csv')
```

```python
features = list(fifa_20.columns[fifa_20.dtypes=='int64'])[11:]
target = ['overall']
```

```python
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
scaler_X.fit(fifa_20[features].values)
scaler_y.fit(fifa_20[target].values)
```

```python
def pre_data(df,scaler_X,scaler_y):
    
    X = df[features]
    y = df[target]
    
    X = scaler_X.transform(X)
    y = scaler_y.transform(y)
    
    return X,y
```

```python
X , y = pre_data(fifa_20,scaler_X,scaler_y)
```

# **Ajuste de un modelo de regresión lineal con pesos no negativos**

## **Visualizando los datos en dos dimenciones utilizando PCA y agrupandolos mediante el algoritmo de K-Means**

```python
pca = PCA()
pca.fit(X)
CPX = pca.transform(X)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(CPX[:,0],CPX[:,1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.plot();
```

```python
kmean = KMeans(n_clusters=4,random_state=1)
kmean.fit(X)
clusters = kmean.predict(X)
```

```python
colors=['red','blue','green','indigo']

plt.figure(figsize=(10,6))

for c,col in zip(np.unique(clusters),colors):
    
    plt.scatter(CPX[clusters==c,0],CPX[clusters==c,1],c=col,label='cluster {}'.format(c))

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.plot();
```

```python
fifa_20['cluster'] = clusters
```

Se identificó que los clusters 0, 1, 2 y 3 corresponden a las posiciones medio, delantero, portero y defensa respectivamente.

```python
fifa_20['position'] = 0

for i in range(fifa_20.shape[0]):
    
    if fifa_20['cluster'].iloc[i] == 0:
        
        fifa_20['position'].iloc[i] = 'MED'
        
    elif fifa_20['cluster'].iloc[i]== 1:
        
        fifa_20['position'].iloc[i] = 'DEL'
        
    elif fifa_20['cluster'].iloc[i] == 2:
        
        fifa_20['position'].iloc[i] = 'POR'
        
    elif fifa_20['cluster'].iloc[i] == 3:
        
        fifa_20['position'].iloc[i] = 'DEF'

# Función que predice la valoración general

def predict_overall(x,model):
    
    x = scaler_X.transform(x)
    
    y = model.predict(x)
    y = scaler_y.inverse_transform(y)
    
    return y
    
```

## **Generando un modelo por cada grupo obtenido con el algoritmo de k-means**




```python
loss = losses.MeanSquaredError()
lr = 0.05
optimizer = optimizers.SGD(learning_rate=lr)
epochs = 100
batch_size = round(X.shape[0]/50)
input_shape = X.shape[1]
n_outputs = 1

constraint = constraints.NonNeg()

callback = [callbacks.EarlyStopping(min_delta=1e-4,patience=3)]
test_split = 0.1

scalers = []
```

## **Porteros**

```python
seed(400)
set_seed(400)
np.random.seed(400)
data_gk = fifa_20[fifa_20['position']=='POR']

scaler_X_gk = MinMaxScaler().fit(data_gk[features].values)
scaler_y_gk = MinMaxScaler().fit(data_gk[target].values)

scalers.append((scaler_X_gk,scaler_y_gk))
X_gk, y_gk = pre_data(data_gk,scaler_X_gk,scaler_y_gk) 

model_gk = single_layer_perceptron(input_shape,n_outputs,optimizer,loss,kernel_constraint=constraint,
                                           activation='linear',use_bias=True) 
history = model_gk.fit(X_gk,y_gk,validation_split=test_split,epochs=epochs,batch_size=batch_size)
```

```python
# Predicción del portero Oblak
x = fifa_20.query('short_name=="J. Oblak"')[features]

predict_overall(x,model_gk)
```

## **Defensas**

```python
seed(400)
set_seed(400)
np.random.seed(400)

data_def = fifa_20[fifa_20['position']=='DEF']

scaler_X_def = MinMaxScaler().fit(data_def[features].values)
scaler_y_def = MinMaxScaler().fit(data_def[target].values)

scalers.append((scaler_X_def,scaler_y_def))
X_def, y_def = pre_data(data_def,scaler_X_def,scaler_y_def) 

model_def = single_layer_perceptron(input_shape,n_outputs,optimizer,loss,kernel_constraint=constraint,
                                           activation='linear',use_bias=True) 

history = model_def.fit(X_def,y_def,validation_split=test_split,epochs=epochs,batch_size=batch_size)
```

```python
x = fifa_20.query('short_name=="K. Koulibaly"')[features]
predict_overall(x,model_def)
```
