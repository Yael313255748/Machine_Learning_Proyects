# **Conjuntos de entrenamiento y prueba**

```python
val_data = train_final.loc[train_final.query('Month == "August"').index,:]
train_data = train_final.loc[train_final.query('Month != "August"').index,:]
```

# **Imputación de datos faltantes**

```python
imputer_cat = SimpleImputer(strategy='most_frequent')
imputer_cat.fit(train_data[categorical_columns])
train_data[categorical_columns] = imputer_cat.transform(train_data[categorical_columns])
val_data[categorical_columns] = imputer_cat.transform(val_data[categorical_columns])
```

```python
numerical_columns = list(train_data.columns[train_data.dtypes==float])

imputer_num = KNNImputer()
imputer_num.fit(train_data[numerical_columns])
train_data[numerical_columns] = imputer_num.transform(train_data[numerical_columns])
val_data[numerical_columns] = imputer_num.transform(val_data[numerical_columns])
```

```python
X_train = train_data.drop(['Credit_Score','Month'],axis=1)
y_train = train_data[['Credit_Score']]

X_val = val_data.drop(['Credit_Score','Month'],axis=1)
y_val = val_data[['Credit_Score']]
```

# **Transformación de las variables**

Se crearon  de variables dummies mediante las categóricas y se escalaron la numéricas con el método *MinMax*.

```python
encoder = OneHotEncoder(sparse=False,drop='first')
scaler = MinMaxScaler()
target_encoder = OneHotEncoder(sparse=False,drop=None)

numeric = X_train.columns[(X_train.dtypes==float) | (X_train.dtypes==int)]
categorical = X_train.columns[(X_train.dtypes==object) ]
transfomers = [('numeric_features',scaler,numeric),
               ('categorical_features',encoder,categorical)]
transfomer = ColumnTransformer(transfomers)
transfomer.fit(X_train)
target_encoder.fit(y_train)
```

```python
X_train_t = transfomer.transform(X_train)
y_train_t = target_encoder.transform(y_train)

X_val_t = transfomer.transform(X_val)
y_val_t = target_encoder.transform(y_val)
```

# **Ajuste de los modelos y búsqueda de los mejores hiperparámetros**

## **Neural Network**

```python
def nn_classifier(lr=0.05,opt='sgd',nn=10,n_hl=2,act='relu',momentum=0.99,
                 epsilon=1e-2):
   '''
    La función genera un modelo de perceptrón multicapa.
    
    Parametros
    ----------
    
    lr : float. Valor de la taza de aprendizaje.
    
    opt : str. Método para optimizar la función de perdida, los valores que puede recibir son: 'sgd', 'adam',
    'RMSprop' y 'Adadelta'.
    
    nn : int. Numero de neuronas por capa oculta.
    
    n_hl : int. Numero de capas ocultas.
    
    act: str. función de activación para las ocultas.
    
    momentum: float. Valor del parámetro momentum para la capa BatchNormalization.
    
    epsilon: float. Valor del parámetro epsilon para la capa BatchNormalization.
    
    Returns
    -------
    
    model : tf.keras.Model. Un modelo de perceptrón multicapa, que además contiene capas de BatchNormalization
    
    '''

    
    n_outputs = 3
    input_shape = (X_train_t.shape[1],)
    
    if opt == 'sgd':
        opt = kr.optimizers.SGD(learning_rate=lr)
        
    elif opt == 'adam':
        opt = kr.optimizers.Adam(learning_rate=lr)
    
    elif opt == 'RMSprop':
        opt = kr.optimizers.RMSprop(learning_rate=lr)
    
    elif opt == 'Adadelta':
        opt = kr.optimizers.Adadelta(learning_rate=lr)
        
    met = [kr.metrics.CategoricalAccuracy()]
    loss = kr.losses.CategoricalCrossentropy()
    out_act = 'sigmoid'
    model = multi_layer_perceptron(input_shape,n_outputs,
                                    nn,n_hl,act,out_act,opt,loss,met,
                                    batch_normalization=True,
                                    momentum_bn=momentum,
                                    epsilon_bn=epsilon)
    
    return model
```

```python
class hyper_model(kt.HyperModel):
    
    def build(self, hp):
        
        lr_space = hp.Float('learning_rate',1e-4,0.1,sampling='log')
        opt_space = hp.Choice('optimizer',['adam','RMSprop'])
        nn_space = hp.Int('number_neurons',2,10,sampling='log')
        hl_space = hp.Int('number_hidden_layers',1,3,sampling='log')
        act_space = hp.Choice('activation',['relu','tanh','elu'])
        mom_space = hp.Float('momentum',1e-4,0.9999,sampling='log')
        eps_space = hp.Float('epsilon',1e-4,0.1,sampling='log')
        model = nn_classifier(lr_space,opt_space,nn_space,hl_space,act_space,mom_space,eps_space)
        
        return model
    
```

```python
hp = kt.HyperParameters()
mlp_hyper_model = hyper_model()
```

```python
# Se utilizó la validacion cruzada con 5 folds 
class CVTuner(kt.Tuner):
    
    def run_trial(self,trial,x,y,batch_size=32,epochs=1,callbacks=None):
        
        cv = KFold(5,shuffle=True,random_state=100)
        val_acc = []

        for train_indices, test_indices in cv.split(x):
            
              x_train, x_test = x[train_indices], x[test_indices]
    
              y_train, y_test = y[train_indices], y[test_indices]
        
              model = self.hypermodel.build(trial.hyperparameters)
              model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs,
                       callbacks=callbacks)
    
              val_acc.append(model.evaluate(x_test, y_test)[1])
        
        return np.mean(val_acc)
    
    
```

```python
# Método bayesiano para la busqueda de los mejores parámetros
oracle = kt.oracles.BayesianOptimization(objective='val_categorical_accuracy',
                                        max_trials=30,
                                        seed=100)
```

```python
cv_tuner = CVTuner(oracle,mlp_hyper_model,directory='hyperparameter_results/',
                  overwrite=True,project_name='Neural_Network')
```

```python
Epochs = 50
Batch_size = 32#round(X_t.shape[0]/50)
Callbacks = [kr.callbacks.EarlyStopping(min_delta=1e-3,patience=3,monitor='categorical_accuracy')]
```

```python
cv_tuner.search(X_train_t,y_train_t,epochs=Epochs,
                batch_size=Batch_size,callbacks=Callbacks)
```

```python
hp_best_top5 = cv_tuner.get_best_hyperparameters(5)
```

```python
i = 0 
for bp in hp_best_top5:
    
    model = hyper_model().build(bp)
    model.save('models/Neural_Network/neural_network_{}.h5'.format(i+1))
    i+=1
    
```

```python
def random_forest_model(hp):
    
    model = RandomForestClassifier(n_estimators=hp.Int('n_estimators',100,1000,sampling='log'),
                                  max_depth=hp.Int('max_depth',2,X_train_t.shape[1],sampling='log'),
                                  criterion=hp.Choice('criterion',['gini','entropy','log_loss']),
                                  random_state=100)
    return model 
```

## **Random Forest**

```python
rf_tuner = kt.tuners.SklearnTuner(
    oracle=kt.oracles.BayesianOptimizationOracle(
        objective=kt.Objective('score', 'max'),
        max_trials=30,
    seed=100),
    hypermodel=random_forest_model,
    scoring=make_scorer(accuracy_score),
    cv=KFold(5,shuffle=True,random_state=100),
    directory='hyperparameter_results/',
    project_name='random_forest',
    overwrite=True
)
```

```python
rf_tuner.search(X_train_t,y_train_t)
```

```python
rf_best_top5 =  rf_tuner.get_best_hyperparameters(5)
i = 0 
for rf in rf_best_top5:
    
    rf_model = random_forest_model(rf)
    pickle.dump(rf_model, open('models/Random_Forest/random_forest_{}'.format(i+1), 'wb'))
    i+=1
    
```

## **Suport Vector Machine**

```python
def svm_model(hp):
    
    model = SVC(C=hp.Float('C',1.0,1000.0,sampling='log'),
               kernel=hp.Choice('kernel',['poly', 'rbf', 'sigmoid']),
               degree=hp.Int('degree',2,3,sampling='log'),
               coef0=hp.Float('coef0',1e-4,1.0,sampling='log'),
               gamma=hp.Float('gamma',1e-4,0.5,sampling='log'))
    return model 

svm_tuner = kt.tuners.SklearnTuner(
    oracle=kt.oracles.BayesianOptimizationOracle(
        objective=kt.Objective('score', 'max'),
        max_trials=30,
    seed=100),
    hypermodel=svm_model,
    scoring=make_scorer(accuracy_score),
    cv=KFold(5,shuffle=True,random_state=100),
    directory='hyperparameter_results/',
    project_name='svm',
    overwrite=True,
)
```

```python
svm_tuner.search(X_train_t,y_train_t.argmax(axis=1))
```

```python
svm_best_top5 =  svm_tuner.get_best_hyperparameters(5)
i = 0 
for svm in svm_best_top5:
    
    svm_model_ = svm_model(svm)
    pickle.dump(svm_model_, open('models/SVM/svm_{}'.format(i+1), 'wb'))
    i+=1
    
```
